Η εκπαίδευση του recommender θα γίνεται ως εξής:

Για κάθε ταινία θα υπάρχει ένα score το όποιο αρχικά θε είναι ίσο με 0 και θα διαμορφώνεται από τις αξιολογήσεις κάθε user. Έτσι για έναν χρήστη με βάση τις αξιολογήσεις που υπάρχουν στο αρχείο train_ratings θα πρέπει:

1.	Για κάθε ταινία που έχει βαθμολογήσει να βρίσκονται οι κοινές ταινίες ανά κλίμακα και στο μέχρι τώρα σκορ κάθε παρόμοιας ταινίας θα προστίθεται ένα βάρος το όποιο θα μπορούσε να είναι το ποσοστό ομοιότητας της ταινίας (δηλαδή ένα βάρος για κάθε κλίμακα) επί τον βαθμό που έχει βάλει ο χρήστης για την αρχική ταινία. Στη συνέχεια, ανάλογα με το σκορ που έχει σχηματιστεί για κάθε ταινία, θα επιλέγεται αν αυτή θα μπορούσε να είναι προτεινόμενη για τον χρήστη.

Η λογική πίσω από την παραπάνω διαδικασία είναι ότι παρόμοιες ταινίες θα έχουν ανάλογο βαθμό. Για παράδειγμα αν ένας χρήστης έχει αξιολογήσει αρκετές  ταινίες οι όποιες είναι sci-fiction με 5/5 τότε μια ταίνια sci-fiction την όποια δεν έχει δεί λογικά θα του αρέσει και θα έπρεπε να την προτείνουμε.

Μετά την εκπαίδευση του συστήματος σας καλείστε να δοκιμάσετε τον recommender που κατασκευάσατε στην πράξη. Για τον σκοπό αυτό θα φορτώσετε το αρχείο train_ratings.csv όπου περιέχονται οι αξιολογήσεις του ίδιου χρήστη για άλλες ταινίες. Το σύστημά σας πρέπει να προβλέπει αν μια ταινία θα πρέπει να προταθεί στον χρήστη. Μια ταίνια θα έπρεπε να έχει προταθεί στον  χρήστη, αν έχει βαθμό μεγαλύτερο του 3. Συνεπώς για την επίβλεψη του συστήματός σας θα πρέπει για κάθε μια από τις ταινίες του αρχείου test_ratings.csv να επιστρέφεται 1 ή 0 αν η ταινία θα έπαιρνε βαθμό μεγαλύτερο του 3 ή όχι, δηλαδή αν θα έπρεπε να την είχαμε προτείνει ή όχι.

Στην συνέχεια, σε συνδιασμό με τις πραγματικές απαντήσεις του χρήστη θα αξιολογήσετε το σύστημά σας χρησιμοποιόντας τις μετρικές: precision, recall, f1 οι οποίες είναι οι πλέον γνωστές μετρικές και ευρέως χρησιμοποιούμενες τεχνικές για την επίβλεψη-μέτρηση απόδοσης ανάλογων συστημάτων.

1. Precision: Δείχνει πόσο ακριβές είναι το σύστημα. Υπολογίζει πόσα από τα στιγμιότυπα τα όποια προβλέψαμε ότι ανήκουν σε μια κλάση όντως ανήκουν σε αυτή. Η μετρική αυτή μας δίνει μια εικόνα σχετικά με τον αριθμό των ταινίων που προβλέψαμε ως προτεινόμενες ενώ δεν θα έπρεπε.

2. Recall: Υπολογίζει πόσα από τα στιγμιότυπα που ανήκουν σε μια κλάση (π.χ. προτεινόμενες ταινίες) προβλέφθηκαν σωστά.

3. F1: Είναι ένας μέσος μεταξύ των παραπάνω δυο μετρικών, έτσι ώστε να διατηρείται μια ισορροπία μεταξύ τους. Υπολογίζεται από την παρακάτω σχέση:

$$F_1=2\frac{Precision\times{Recall}}{Precision+Recall}$$

````python
import json
import os
import pandas as pd
from pyswip import Prolog
#Στον κώδικα της pyswip στο αρχείο core.py αλλάζω στη γραμμή 180 το r'pl\bin' σε r'swipl\bin'
#Αρχείο C:\Users\<USERNAME>\AppData\Local\Programs\Python\Python38\Lib\site-packages\pyswip
````

````python
# Η βιβλιοθήκη pandas είναι χρήσιμη για την εργασία με τέτοια δεδομένα
import pandas as pd
path=''
# Διάβασμα του αρχείου 'movie_metadata.csv' 
data = pd.read_csv(path + "movies_metadata.csv") 
#Στο csv υπαρχούν κελία με nan τιμές
#Στις θέσεις αυτές βάζουμε 'UNK' πράγμα που  κάνουμε με την παρακάτω συνάρτηση
data.fillna("UNK", inplace=True)
# Preview the first 5 lines of the loaded data 
#data.head()
````

````python
def clean_text(text):
  text = text.replace(u'\xa0', u'')
  text = text.replace(u"'", u'')
  return text

def clean_data(text):
  text = text.replace(u'"', u"'")
  text = text.replace(u"'name': '", u'"name": "')
  text = text.replace(u"'name'", u'"name"')
  text = text.replace(u"', 'id'", u'", "id"')
  text = text.replace(u"'id'", u'"id"')
  return text
````

````python
#create World
#Ορίζουμε τον κόσμο μας
prolog = Prolog()

if not os.path.isfile('db.pl'):
  #Για κάθε row του πίνακα φτιάχνουμε τα κατηγορήματα που θέλουμε να αποθηκέυσουμε
  #αρχικά σε μια λίστα με το όνομα literals
  literals = []
  for row in data.itertuples(index=True, name='Pandas'):
    #Get movie title
    movie_title = clean_text(getattr(row, 'movie_title'))

    #Get genre predicates
    for genre in getattr(row, 'genres').split("|"):
      literals.append("genre('"+ movie_title +"','"+ genre +"')")

    #Get plotKeyword predicates
    for plotKeyword in getattr(row, 'plot_keywords').split("|"):
      literals.append("plotKeyword('"+ movie_title +"','"+ clean_text(plotKeyword) +"')")
    
    #Get language predicates
    literals.append("language('"+ movie_title +"','"+ clean_text(getattr(row, 'language')) +"')")

    #Get popularity predicates
    if getattr(row, 'popularity') != 'UNK':
      literals.append("popularity('"+ movie_title +"',"+ str(round(float(getattr(row, 'popularity')), 3)) +")")

    #Get productionCompany predicates
    for productionCompany in json.loads(clean_data(getattr(row, 'production_companies'))):
      literals.append("productionCompany('"+ movie_title +"',"+ str(productionCompany["id"]) +")")
    
    #Get grossRevenue predicates
    if getattr(row, 'gross') != 'UNK':
      literals.append("grossRevenue('"+ movie_title +"',"+ str(getattr(row, 'gross')) +")")

    #Get duration predicates
    if getattr(row, 'duration') != 'UNK':
      literals.append("duration('"+ movie_title +"',"+ str(getattr(row, 'duration')) +")")

    #Get voteAverage predicates
    if getattr(row, 'vote_average') != 'UNK':
      literals.append("voteAverage('"+ movie_title +"',"+ str(getattr(row, 'vote_average')) +")")

    #Get usersVoted predicates
    if getattr(row, 'num_voted_users') != 'UNK':
      literals.append("usersVoted('"+ movie_title +"',"+ str(getattr(row, 'num_voted_users')) +")")

    #Get releaseYear
    if getattr(row, 'title_year') != 'UNK':
      literals.append("releaseYear('"+ movie_title +"',"+ str(getattr(row, 'title_year')) +")")

    #Get country predicates
    literals.append("country('"+ movie_title +"','"+ getattr(row, 'country') +"')")

    #Get director predicates
    literals.append("director('"+ movie_title +"','"+ clean_text(getattr(row, 'director_name')) +"')")

    #Get playsIn predicates
    literals.append("playsIn('"+ movie_title +"','"+ clean_text(getattr(row, 'actor_1_name')) +"')")
    literals.append("playsIn('"+ movie_title +"','"+ clean_text(getattr(row, 'actor_2_name')) +"')")
    literals.append("playsIn('"+ movie_title +"','"+ clean_text(getattr(row, 'actor_3_name')) +"')")


  #Create file to store predicates
  f = open("db.pl", "w", encoding="utf-8")

  #Η Prolog θέλει τα κατηγορήματά της με την σειρά
  literals.sort()
  for literal in literals:
    prolog.assertz(literal)
    f.write(literal + ".\n")

  f.close()

else:
  #Επίσης μπορούμε να κάνουμε consult με ένα έτοιμο αρχείο τον κόσμο μας, όπως παρακάτω
  prolog.consult("db.pl")
````

````python
'''
Απλοί Κανόνες
'''
if not os.path.isfile('rules.pl'):
    rules = []
    f = open("rules.pl", "w", encoding="utf-8")
    #Κοινό θέμα (4 κοινές λέξεις genre)
    rules.append('''extraCommonGenre(X,Y) :- findall(M1, genre(X,M1),Z1),!,bagof(M2, genre(Y,M2),Z2),
                                             intersection(Z1, Z2, Z),
                                             length(Z, W), W >= 4, X \= Y''')

    #Αρκετά κοινό θέμα (3 κοινές λέξεις genre)
    rules.append('''veryCommonGenre(X,Y) :- findall(M1, genre(X,M1),Z1),!,bagof(M2, genre(Y,M2),Z2),
                                            intersection(Z1, Z2, Z),
                                            length(Z, W), W >= 3, X \= Y''')

    #Σχετικά κοινό θέμα (2 κοινές λέξεις genre)
    rules.append('''moreCommonGenre(X,Y) :- findall(M1, genre(X,M1),Z1),!,bagof(M2, genre(Y,M2),Z2),
                                            intersection(Z1, Z2, Z),
                                            length(Z, W), W >= 2, X \= Y''')

    #Λίγο κοινό θέμα (1 κοινές λέξεις genre)
    rules.append('''lessCommonGenre(X,Y) :- findall(M1, genre(X,M1),Z1),!,bagof(M2, genre(Y,M2),Z2),
                                            intersection(Z1, Z2, Z),
                                            length(Z, W), W >= 1, X \= Y''')

    #Ακριβώς ίδια πλοκή (7 κοινές λέξεις plotKeyword)
    rules.append('''veryCommonPlot(X,Y) :- findall(M1, plotKeyword(X,M1),Z1),!,bagof(M2, plotKeyword(Y,M2),Z2),
                                           intersection(Z1, Z2, Z),
                                           length(Z, W), W >= 7, X \= Y''')

    #Σχετικά ίδια πλοκή (5 κοινές λέξεις plotKeyword)
    rules.append('''moreCommonPlot(X,Y) :- findall(M1, plotKeyword(X,M1),Z1),!,bagof(M2, plotKeyword(Y,M2),Z2),
                                           intersection(Z1, Z2, Z),
                                           length(Z, W), W >= 5, X \= Y''')

    #Λίγο ίδια πλοκή (3 κοινές λέξεις plotKeyword)
    rules.append('''lessCommonPlot(X,Y) :- findall(M1, plotKeyword(X,M1),Z1),!,bagof(M2, plotKeyword(Y,M2),Z2),
                                           intersection(Z1, Z2, Z),
                                           length(Z, W), W >= 3, X \= Y''')

    #Ίδια γλώσσα
    rules.append('''commonLanguage(X,Y) :- language(X,Z), language(Y,Z), X \= Y''')

    #Θρυλικό επίπεδο φήμης (χρήση  popularity, voteAverage, usersVoted, grossRevenue)
    rules.append('''legendaryFame(X) :- popularity(X, M1), M1 >= 90, 
                                        voteAverage(X, M2), M2 >= 8,
                                        usersVoted(X, M3), M3 >= 5000,
                                        grossRevenue(X, M4), M4 >= 50000000''')
    #Υψηλό επίπεδο φήμης
    rules.append('''greatFame(X) :- popularity(X, M1), M1 >= 65,
                                    voteAverage(X, M2), M2 >= 6.8,
                                    usersVoted(X, M3), M3 >= 3500,
                                    grossRevenue(X, M4), M4 >= 15000000, not(legendaryFame(X))''')

    #Μέτριο επίπεδο φήμης
    rules.append('''mediumFame(X) :- popularity(X, M1), M1 >= 40,
                                     voteAverage(X, M2), M2 >= 5.5,
                                     usersVoted(X, M3), M3 >= 1000,
                                     grossRevenue(X, M4), M4 >= 5000000,
                                     not(legendaryFame(X)), not(greatFame(X))''')

    #Χαμηλό επίπεδο φήμης
    rules.append('''lowFame(X) :- findall(M1, legendaryFame(M1), Z1),
                                  findall(M2, greatFame(M2), Z2),
                                  findall(M3, mediumFame(M3), Z3),
                                  union(Z1, Z2, Z4),union(Z4, Z3, Z5),
                                  findall(M, director(M,_), Z6),
                                  list_to_set(Z6, Z7), subtract(Z7, Z5, Z),!,
                                  member(X,Z)''')
    #Ίδια Φήμη
    rules.append('''sameFame(X,Y) :- ((legendaryFame(X), legendaryFame(Y)) ;
                                       (greatFame(X), greatFame(Y)) ;
                                       (mediumFame(X), mediumFame(Y)) ;
                                       (lowFame(X), lowFame(Y))), X \= Y''')

    #Πολύ Κοινή παραγωγή (4 κοινά productionCompany)
    rules.append('''veryCommonProduction(X,Y) :- findall(M1, productionCompany(X,M1),Z1),!,bagof(M2, productionCompany(Y,M2),Z2),
                                                 intersection(Z1, Z2, Z),
                                                 length(Z, W), W >= 4, X \= Y''')

    #Σχετικά κοινή παραγωγή (2 κοινά productionCompany)
    rules.append('''moreCommonProduction(X,Y) :- findall(M1, productionCompany(X,M1),Z1),!,bagof(M2, productionCompany(Y,M2),Z2),
                                                 intersection(Z1, Z2, Z),
                                                 length(Z, W), W >= 2, X \= Y''')

    #Λίγο κοινή παραγωγή (1 κοινό productionCompany)
    rules.append('''lessCommonProduction(X,Y) :- findall(M1, productionCompany(X,M1),Z1),!,bagof(M2, productionCompany(Y,M2),Z2),
                                                 intersection(Z1, Z2, Z),
                                                 length(Z, W), W >= 1, X \= Y''')

    #Μικρής διάρκειας ταινία (μέχρι 40 λεπτά)
    rules.append('''shortFilm(X) :- duration(X, Y), Y =< 40''')

    #Μεσαίας διάρκειας ταινία (40 με 90 λεπτά)
    rules.append('''mediumFilm(X) :- duration(X, Y), Y > 40, Y =< 90''')

    #Μεγάλης διάρκειας ταινία (90+ λεπτά)
    rules.append('''largeFilm(X) :- duration(X, Y), Y > 90''')

    #Ίδιας διάρκειας ταινια
    rules.append('''sameDuration(X,Y) :- ((shortFilm(X), shortFilm(Y)) ;
                                          (mediumFilm(X), mediumFilm(Y)) ;
                                          (largeFilm(X), largeFilm(Y))), X \= Y''')

    #Πολύ παλιά (releaseDate 1900-1935)
    rules.append('''veryOldFilm(X) :- releaseYear(X, Y), Y =< 1935''')

    #Λιγότερο παλιά (releaseDate 1935-1960)
    rules.append('''oldFilm(X) :- releaseYear(X, Y), Y > 1935, Y =< 1960''')

    #Ενδιάμεσα (releaseDate 1960-1990)
    rules.append('''intermediateFilm(X) :- releaseYear(X, Y), Y > 1960, Y =< 1990''')

    #Πρόσφατη (releaseDate 1990-2010)
    rules.append('''newFilm(X) :- releaseYear(X, Y), Y > 1990, Y =< 2010''')

    #Πολύ πρόσφατη (releaseDate 2010+)
    rules.append('''veryNewFilm(X) :- releaseYear(X, Y), Y > 2010''')

    #Ίδια εποχή
    rules.append('''sameEra(X,Y) :- ((veryOldFilm(X), veryOldFilm(Y)) ;
                                     (oldFilm(X), oldFilm(Y)) ;
                                     (intermediateFilm(X), intermediateFilm(Y)) ;
                                     (newFilm(X), newFilm(Y)) ;
                                     (veryNewFilm(X), veryNewFilm(Y))), X \= Y''')

    #Ίδια χώρα
    rules.append('''commonCountry(X,Y) :- country(X,Z), country(Y,Z), X \= Y''')

    #Κοινός Σκηνοθέτης
    rules.append('''commonDirector(X,Y) :- director(X,Z), director(Y,Z), X \= Y''')

    #Ίδιους τους βασικούς ηθοποιούς (3 κοινά playsIn)
    rules.append('''sameStarActors(X,Y) :- findall(M1, playsIn(X,M1),Z1),!,bagof(M2, playsIn(Y,M2),Z2),
                                           intersection(Z1, Z2, Z),
                                           length(Z, W), W =:= 3, X \= Y''')

    #Αρκετά ίδιους βασικούς ηθοποιούς (2 κοινά playsIn)
    rules.append('''moreStarActors(X,Y) :- findall(M1, playsIn(X,M1),Z1),!,bagof(M2, playsIn(Y,M2),Z2),
                                           intersection(Z1, Z2, Z),
                                           length(Z, W), W >= 2, X \= Y''')

    #Σχετικά ίδιους ηθοποιούς (1 κοινό playsIn)
    rules.append('''lessStarActors(X,Y) :- findall(M1, playsIn(X,M1),Z1),!,bagof(M2, playsIn(Y,M2),Z2),
                                           intersection(Z1, Z2, Z),
                                           length(Z, W), W >= 1, X \= Y''')

    #Είναι έγχρωμες ή ασπρόμαυρες
    rules.append('''black_and_white(X) :- plotKeyword(X, 'black and white') ; plotKeyword(X, 'black and white scene') ; 
                                          (releaseYear(X, T), T =< 1950)''')

    rules.append('''color(X) :- findall(M1, plotKeyword(M1, 'black and white'), Z1), 
                                findall(M2, plotKeyword(M2, 'black and white scene'), Z2), 
                                union(Z1, Z2, Z3), findall(M3, plotKeyword(M3,_), Z4),
                                list_to_set(Z4, Z5), subtract(Z5, Z3, Z),!, 
                                member(X, Z), releaseYear(X, T), T > 1950''')

    rules.append('''sameColor(X,Y) :- (color(X), color(Y)) ;
                                      (black_and_white(X), black_and_white(Y)),
                                      X \= Y''')

    for rule in rules:
        prolog.assertz('('+ rule +')')
        f.write(rule + ".\n\n")
    f.close()
else:
  #Επίσης μπορούμε να κάνουμε consult με ένα έτοιμο αρχείο τους κανόνες μας, όπως παρακάτω
  prolog.consult("rules.pl")
````

````python
def simple_recommender(movie, sim):
    s = set()
    q = prolog.query("find_sim_"+ str(sim) +"('" + movie +"',M)")
    for soln in q:
        m = soln['M'] 
        if m not in s:
            s.add(soln['M'])
    q.close()
    answers = s
    return answers
````

# **3ο Μέρος: Recommendation System Με βάση τις προτιμήσεις-Αξιολογήσεις του χρήστη-Εκπαίδευση και Πρόβλεψη**


Αρχικά μελετάμε τις αξιολογήσεις κάθε χρήστη για να καταλάβουμε την δομή και τις πληροφορίες κάθε αρχείου.




````python
from tqdm.notebook import tqdm
from sklearn.metrics import precision_score, recall_score, f1_score

#ratings_weights δείχνουν την πιθανότητα μια ταινία να προταθεί(+) ή να αντι-προταθεί(-) σε σύνολο 10 ατόμων
#π.χ. για 3.5 rating θα προταθεί σε 6/10 άτομα, ενώ για 0.5 rating θα αντι-προταθεί σε 8/10 άτομα
rating_weights = {0.0: -1, 0.5: -0.8, 1.0: -0.6, 1.5: -0.4, 2.0: 0, 2.5:0.2, 3.0:0.4, 3.5: 0.6, 4.0:0.8, 4.5: 0.9, 5.0:1}

#scoring_weights δείνουν βάρη στις όμοιες ταινίες (σηματοδοτούν σε πόσες 10άδες θα γίνει η πρόταση ή η αντι-πρόταση)
#Το σύστημα για τις κατηγοορίες find_sim_2 μέχρι find_sim_5 βρίσκει πολύ όμοιες ταινίες άρα θα αυξήσουμε τα βάρη ανάλογα
score_weights = {0:0.8, 1: 1.4, 2: 2, 3: 2.5, 4: 4.5, 5: 6}


def train_recommender(ratings, rating_weights, score_weights, start=0, end = -1, movie_score = {}):
    """
    Στην συνάρτηση αυτή μπορούμε να ορίζουμε ποιο υποσύνολο των αξιολογήσεων θα χρησιμοποιήσουμε για το train,
    ενώ παράλληλα μπορούμε να αρχικοποιήσουμε και το movie_score το οποίο αποθηκεύονται τα score κάθε ταινίας.
    Έτσι αν έχουμε κάνει train τον recommender μας για τις πρώτες 10 ταινίες, μπορούμε να συνεχίσουμε για τις 
    υπόλοιπες χωρίς κάθε φορά να ξανά-διαμορφώνουμε το score για αυτές. Αυτό θα σας βοηθήσει στην εκτέλεση και χρονικά όταν ελέγχετε την 
    απόδοση σε υποσύνολα των ratings π.χ. πρώτες 10 ταινίες έπειτα τις επόμενες 20 κ.ο.κ.
    """
    if end == -1:
        end = len(ratings)
    ratings = ratings.iloc[range(start, end)]
    for row in tqdm(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')
        already_sim_movies = set()

        for sim in range(5, -1, -1):
          #Find simlar movies in specific similarity class
          q = prolog.query("sim_"+ str(sim) +"('"+ movie +"', Y)")
          similar_movies = set()
          for soln in q:
            m = soln['Y'] 
            if m not in similar_movies:
              similar_movies.add(soln['Y'])

          #Change movie_score
          for similar_movie in similar_movies:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[rating] * score_weights[sim]
                already_sim_movies.add(similar_movie)
            elif similar_movie not in already_sim_movies:
                movie_score[similar_movie] += rating_weights[rating] * score_weights[sim] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος
    
    return movie_score


#αυτό είναι ένα παράδειγμα για το πως θα μπορούσε να υλοποιήθει η predict
#έχουμε ορίσει ότι μια ταινία θα έπρεπε να είναι προτεινόμενη αν είχε σκορ > 0
def predict_example(ratings, movie_score):
    real, pred = [], []
    for i, row in enumerate(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        if movie in movie_score: #αν έχουμε σχηματίσει βαθμολογία για την ταινία αυτή
            pred.append(int(movie_score[movie] > 1.2)) #heuristic για το αν μια ταινία είναι προτεινόμενη
            real.append(int(rating > 3))# έτσι ορίζουμε ότι μια ταινία θα έπρεπε να είναι προτεινόμενη
            #η συνθήκη αυτή δεν μπορεί να αλλάξει
        else: #δεν μπορούμε να προτείνουμε κάτι για το όποιο δεν έχουμε σχηματίσει εικόνα
            pred.append(0)
            real.append(int(rating > 3))

    return real, pred


def get_metrics(real, pred):
    metrics = {}
    metrics["precision"] = precision_score(real, pred)
    metrics["recall"] = recall_score(real, pred)
    metrics["f1"] = f1_score(real, pred)
    return metrics
````
